```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import precision_score,confusion_matrix,accuracy_score,recall_score,classification_report, f1_score
```

```python
data=pd.read_csv("Iris.csv")
data
```

```python
data=pd.read_csv("Iris.csv")
encoder=preprocessing.LabelEncoder()
data['Species']=encoder.fit_transform(data['Species'])
data
```

```python
x=data.iloc[:,[1,4]].values
y=data.iloc[:,5].values

x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.20, random_state=0)

classifier=GaussianNB()
classifier.fit(x_train,y_train)

y_pred=classifier.predict(x_test)
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)

acc=accuracy_score(y_test,y_pred)
print(acc)

pr=precision_score(y_test,y_pred,average=None)
pr

recall=recall_score(y_test,y_pred,average=None)
recall

cl_report=classification_report(y_test,y_pred)
cl_report

f1=f1_score(y_test,y_pred,average=None)
print(f1)
```

```python
Introduction to Naive Bayes Classifier:
The Naive Bayes classifier is a widely used machine learning algorithm based on
Bayes' theorem. It is known for its simplicity and effectiveness in various
applications, such as text classification and spam filtering. One of its main
advantages is its efficiency with large datasets.
Theory Behind Naive Bayes Classifier:
At its core, the Naive Bayes classifier relies on Bayes' theorem, which calculates the
probability of an event based on prior knowledge of conditions related to the event.
The "naive" part of Naive Bayes stems from the assumption of independence
between predictors. This means that the presence of one feature does not influence
the presence of another. Although this assumption is simplistic, Naive Bayes often
performs well in practice.
Example:
Naive Bayes Classifier on Iris Dataset:
Let's illustrate the Naive Bayes classifier using the Iris dataset, which contains
information about iris flowers' attributes and species.
1. Load the Dataset:
Begin by loading the Iris dataset, typically available in libraries like scikit-learn.
2. Preprocess the Data:
Preprocess the dataset by separating the features (e.g., sepal length, sepal width,
petal length, petal width) and the target variable (species).
3. Train the Naive Bayes Classifier:
Train the Naive Bayes classifier using the training data. In scikit-learn, you can use
classes like GaussianNB for continuous features or MultinomialNB for discrete
features.
4. Make Predictions:
Once trained, use the classifier to predict the species of new iris flowers based on
their features. The classifier selects the species with the highest probability.
5. Evaluate the Model:
Assess the model's performance using metrics like accuracy, precision, recall, or
F1-score on a separate test dataset.
Conclusion:
In conclusion, the Naive Bayes classifier is a straightforward yet powerful algorithm
for classification tasks. Despite its assumption of feature independence, it often
performs well in practice, especially with high-dimensional data. It is efficient, easy to
implement, and suitable for large datasets. However, its performance may suffer if
the independence assumption is violated or if there are strong correlations between
features. Overall, Naive Bayes is a valuable tool in the arsenal of data scientists for
various classification problems.
```
